```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mutual_info_score, accuracy_score, roc_curve, auc, roc_auc_score
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression

from IPython.display import display
import pickle
import requests
import math
```

```python
#using pandas we enforce every string to be numeric, except for empty space we put NAN
total_charges = pd.to_numeric(df.TotalCharges, errors='coerce')
#using this function to check that our dataframe has spaces in totalcharges, related to some customer ids
df[total_charges.isnull()][['customerID', 'TotalCharges']]
```

```python
# we convert to numeric and replace any space with NAN
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
#fill missing number with zeros (NAN will be zero)
df.TotalCharges = df.TotalCharges.fillna(0)
```

```python
#Data cleaning.

# smalling all letters then replace spaces for underscores.
df.columns = df.columns.str.lower().str.replace(' ','_')
#select a subset of datatypes of values that are strings and return their indecies.
string_columns = list(df.dtypes[df.dtypes == 'object'].index)
# loop over indcies to lowercase them and replace space with underscores also.
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
#convert yes to True and no to False then casting to int to make them 1 and 0, to be able to use them for binary classification
df.churn = (df.churn == 'yes').astype(int)
```

```python
#preparing dataframes to BE relaible to make EDA
#model_selection from Scikit-Learn library can handle data splitting by shuffling then split.
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)
#setting the random seed (random_state=1) is important to be sure that our result is reproducible
df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=11)

#Take the column with the target variable churn and save it outside the dataframe.
y_train = df_train.churn.values
y_val = df_val.churn.values

#Delete the churn columns from both the dataframes to make sure we donâ€™t accidentally use the churn variable as a
#feature during training.
del df_train['churn']
del df_val['churn']

# EDA(exploratory data analysis) => to identify the most important features of our data
# As The more we know about the data and the problems inside, the better model we can build afterward.

# 1- check for any missing values in the dataset
# by checking for null values and them them for each column, this process is elementwise
df_train_full.isnull().sum()

# 2- check the distribution of values in the target variable
df_train_full.churn.value_counts()

# 3- check the proportion of churned users or the probability that a customer will churn
global_mean = df_train_full.churn.mean()
# we have a class imblanace as there are three times more people who didn't churn in our data than who did churn
round(global_mean, 3) # 27% of our custmers stopped using our services

# we have both categorical and numerical data, which we needs to deal wtih them separetly.
categorical = ['gender', 'seniorcitizen', 'partner', 'dependents',
                'phoneservice', 'multiplelines', 'internetservice',
                'onlinesecurity', 'onlinebackup', 'deviceprotection',
                'techsupport', 'streamingtv', 'streamingmovies',
                'contract', 'paperlessbilling', 'paymentmethod']

numerical = ['tenure', 'monthlycharges', 'totalcharges']

#see how many unique variables each variable has
df_train_full[categorical].nunique()

# 4- feature importance analysis: understanding which features may be important for our model.
# by Knowing how other variables affect the target variable, churn
female_mean = df_train_full[df_train_full.gender == 'female'].churn.mean()
print('gender == female:', round(female_mean, 3))

# difference between male_churn rate and global churn rate is small, which means that gender is not very useful variable when predicting churn, this also applied for female.
male_mean = df_train_full[df_train_full.gender == 'male'].churn.mean()
print('gender == male:', round(male_mean, 3))

partner_yes = df_train_full[df_train_full.partner == 'yes'].churn.mean()
print('partner == partner_yes:', round(partner_yes, 3))
# the ones who don't have partners (don't live with someone), have a higher probaboility to churn of 33%
# which indicates that the partner variable is useful for predicting churn.
partner_no = df_train_full[df_train_full.partner == 'no'].churn.mean()
print('partner == partner_no:', round(partner_no, 3))

# RISK Ratio: the ratio between probabilities in different groups, where risk refers to the risk of having the effect, which is here (churn customer) compared with the entire population
# note: a group with risk close to 1 is not risky at all

#calculating mean for churn for each column related to the global churn rate
#then, getting differences and compare rates of each column to the global rate to see what categories have a higher effect churn rate
#just by looking at the differences and the risks, we can identify the most discriminative features: the features that are helpful for detecting churn.
for col in categorical:
    df_group = df_train_full.groupby(by=col).churn.agg(['mean']) #using pandas sql-like functions
    df_group['diff'] = df_group['mean'] - global_mean
    df_group['risk'] = df_group['mean'] / global_mean
    display(df_group) # we can't display dataframes inside a loop, so we called display by importing it above

# mutual information: tells us how much information we learn about one variable if we get to learn the value of the other variable
# if the mutual information between a categorical variable and the target is high, this categorical variable will
# be quite useful for predicting the target.
def calculate_mi(series):
    return mutual_info_score(series, df_train_full.churn)

#apply calculate_mi function on the categorical values of df_train_fullint
df_mi = df_train_full[categorical].apply(calculate_mi)
#we sort then convert from pandas series(col) to a dataframe(row,col)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')
df_mi

# mutual information can quantify the degree of dependency between two categorical variables
#but it doesn't work with numerical features, instead we will use "Correlation Coeffcient" with numerical features.
df_train_full[numerical].corrwith(df_train_full.churn)

# 5- feature engineering: we need to transforming all categorical variables to numeric features.
#to be able to train model on them, as model understands numbers in matrices.

#using one-hot encoding technique.
# convert our dataframe to a list of dictionaries
train_dict = df_train[categorical + numerical].to_dict('records')

# The fit method looks at the content of these dictionaries and figures out the
# possible values for each variable and how to map them to the columns in the output matrix.
# DictVectorizer chnages the name of each categorical variable and assign to it the right numerical value, while keping the numerical variables in it's original names without any change
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)

# our transformed matrix
x_train = dv.transform(train_dict)
x_train[0]

dv.get_feature_names()

# 6- Train Our Model (classification model: logistic regression)
# logistic regression is linear cause it uses the dot product

#translates our math to python: for clarification only
# def logistic_regression(xi):
#     # We use score to mean the intermediate result before applying the sigmoid function
#     score = bias
#     for j in range(n):
#         score = score + xi[j] * w[j]
#     prob = sigmoid(score)
#     return prob

# # the output of a logistic regression is probability, so we will use sigmoid function for that purpose.
# def sigmoid(score):
#     # The probability is the result of applying the sigmoid function to the
#     # score; this is the final output, and it can take only the values between 0 and 1.
#     return 1 / (1 + math.exp(-score))

# start logistic regression training using Scikit-Learn
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(x_train, y_train)

val_dict = df_val[categorical + numerical].to_dict(orient='rows') #we fitted the data before so no need to perform it again.
x_val = dv.transform(val_dict)

# get values from the second column of probability(soft predictions) as two columns are conveying the same information
y_pred = model.predict_proba(x_val)[:,1] #produces elementwise y_pred array

# To get the binary predictions(True or False) we take the probabilities and cut them above a certain threshold(comparison underneth
churn = y_pred >= 0.5 #comparison operator is applied elementWise(it's applied for each element) in numpy

# the resulting number of the mean tells us the percentage of correct predictions (accuracy of the model).
(y_val == churn).mean()

# Model interpretation (see how we can interpret the coefficients the model learned.)
# the bias term
model.intercept_[0]

# To see which feature is associated with each weight
dict(zip(dv.get_feature_names(), model.coef_[0].round(3)))

# 7- To build the intuition, letâ€™s train a simpler and smaller model that uses only three variables: contract, tenure, and total charges.
small_subset = ['contract', 'tenure', 'totalcharges']
train_dict_small = df_train[small_subset].to_dict('records')
dv_small = DictVectorizer(sparse=False) #applying one-hot encoding
dv_small.fit(train_dict_small)

x_small_train = dv_small.transform(train_dict_small)
dv_small.get_feature_names()

model_small = LogisticRegression(solver='liblinear', random_state=1)
model_small.fit(x_small_train, y_train)

# bias term
model_small.intercept_[0]

dict(zip(dv_small.get_feature_names(), model_small.coef_[0].round(3)))

# 8- Using the model

# When we prepare items for prediction, they should undergo the same preprocessing steps we did for training the model.
# customer_2 has a very low probability of 7% that it will churn (thus true related to our analysis, as it is a yearly-contract customer and have a big tenture number)
customer_1 = {
    'customerid': '8879-zkjof',
    'gender': 'female',
    'seniorcitizen': 0,
    'partner': 'no',
    'dependents': 'no',
    'tenure': 41,
    'phoneservice': 'yes',
    'multiplelines': 'no',
    'internetservice': 'dsl',
    'onlinesecurity': 'yes',
    'onlinebackup': 'no',
    'deviceprotection': 'yes',
    'techsupport': 'yes',
    'streamingtv': 'yes',
    'streamingmovies': 'yes',
    'contract': 'one_year',
    'paperlessbilling': 'yes',
    'paymentmethod': 'bank_transfer_(automatic)',
    'monthlycharges': 79.85,
    'totalcharges': 3320.75,
}

# customer_2 has a high probability of 83% that it will churn (as it is a month-to-month customer, with a tenture= 1 month)
customer_2 = {
'gender': 'female',
'seniorcitizen': 1,
'partner': 'no',
'dependents': 'no',
'phoneservice': 'yes',
'multiplelines': 'yes',
'internetservice': 'fiber_optic',
'onlinesecurity': 'no',
'onlinebackup': 'no',
'deviceprotection': 'no',
'techsupport': 'no',
'streamingtv': 'yes',
'streamingmovies': 'no',
'contract': 'month-to-month',
'paperlessbilling': 'yes',
'paymentmethod': 'electronic_check',
'tenure': 1,
'monthlycharges': 85.7,
'totalcharges': 85.7
}

x_test = dv.transform([customer_2])
model.predict_proba(x_test)[0, 1]

# 9- Evaluating a binary classification model using: confusion table, precision and recall, the ROC(reciever operating characteristics) and AUC(area under the ROC curve)
# Evaluating means: whether 80% accuracy is good or not and what it actually means in terms of the quality of our model.

# Decide on which threshold is the best for our model 
thresholds = np.linspace(0, 1, 11)
#looping over thresold numpy array and compare it to our actual 
for t in thresholds:
    churn = y_pred >= t
    acc = accuracy_score(y_val, churn) #using skit-learn to calculate accuracy score, that we calculated previously above manually
    print('%0.2f %0.3f' % (t, acc))
    
# using Matplotlib and create a plot that shows how accuracy changes depending on the threshold.
thresholds = np.linspace(0, 1, 21) 
accuracies = [] 
for t in thresholds:
    acc = accuracy_score(y_val, y_pred >= t) 
    accuracies.append(acc) 

plt.plot(thresholds, accuracies)

#depending on the model_small we trained above 
val_dict_small = df_val[small_subset].to_dict('records')

x_small_val = dv_small.transform(val_dict_small)
y_pred_small = model_small.predict_proba(x_small_val)[:,1]

churn_small = y_pred_small >= 0.5
#small model is 76%, so our big model is better with 4% only,  this still doesn't till us if 80% is good or not
accuracy_score(y_val, churn_small) 

# creating a Dummy Baseline: model that predicts the same value(related to the most frequent results in our dataset due to class imbalance, which is not-churned users, so it will be always false), regardless of the features i thas.
size_val = len(y_val) #A
baseline = np.repeat(False, size_val)# numpy array of repeated values related to the number of customers in the dataset

# the naive baseline is 74% , the small model is 2% better and large model is 6% better
#if we think about the trouble we have gone through to train this large model 6% doesn't seem like a significant improvement over the naive baseline model
accuracy_score(baseline, y_val)

# using Inferential statistics to apply the "confusion table", which will clear our mind more about the accuracy of our model
# specifically, we will use Frequentism statistics, specifically, p-values (take in mind that p-value Distributions depend on the statistical power)
# why we use p-values here => when the mean difference or the test statistic is one of the two tails of the normal distribution(here to have true or false)
# here we have four values as a result => False-Flase, True-True, False-True(type 1 error "false positive"), True-False (type 2 error "true negative")
# using p-value we try to cotrol the errors when we draw inferences from our data, by controlling the amount of times in the long run that we make a mistake
# note that, we need large samples to statistically Detect a small effect.

t = 0.5 #threshold
predict_churn = (y_pred >= t)
predict_no_churn = (y_pred < t)

actual_churn = (y_val == 1)
actual_no_churn = (y_val == 0)

true_positive = (predict_churn & actual_churn).sum()
false_positive = (predict_churn & actual_no_churn).sum()

false_negative = (predict_no_churn & actual_churn).sum()
true_negative = (predict_no_churn & actual_no_churn).sum()

confusion_table = np.array(
[[true_negative, false_positive],
[false_negative, true_positive]])

confusion_table / confusion_table.sum()

#calculate accuracy based on confusion_table correct predictions (true_negative and true_positive)
accuracy = (true_negative + true_positive) / (true_negative + true_positive + false_positive + false_negative)

# Precision and recall(based on confusion_table): helps us understand the quality of the model in cases of class imbalance 
# Precision and recall are very useful metrics, but they describe the performance of a classifier only at a certain threshold. which is here 0.5

#first, we calculate "percision" based on correctly predicted positive examples (how many of the positive predictions turned out to be correct)
# Precision helps us understand how many people received a promotional message by mistake. here is 38%
percision = true_positive / (true_positive + false_positive)

#second, we calculate "Recall": is the fraction of correctly classified positive examples among all positive examples.
# Recall helps us understand how many of the churning customers we failed to find. here is 41%
recall = true_positive / (true_positive + false_negative)

# Often we need a metric that summarizes the performance of a classifier for all possible threshold choices

#ROC(receiver operating characteristics)curve: it shows how well a model can separate two classes, positive(churn) and negative(no-churn).
#we need to calculate two quantites FPR and TPR

# FPR(false_positive rate): The fraction of false positives among all negative examples
# FPR is the fraction of users we predicted as churning among everybody who didnâ€™t churn.
FBR = false_positive / (false_positive + true_negative)

# TPR(true positive rate): The fraction of true positives among all positive examples
# TPR is the fraction of users who we predicted as churning among everybody who actually did churn.
TPR = true_positive / (true_positive + false_negative)

# Computing the confusion table for different threshold
def tpr_fpr_dataframe(y_val, y_pred):
    scores = [] 
    
    thresholds = np.linspace(0, 1, 101) 
    
    for t in thresholds: 
        tp = ((y_pred >= t) & (y_val == 1)).sum() 
        fp = ((y_pred >= t) & (y_val == 0)).sum() 
        fn = ((y_pred < t) & (y_val == 1)).sum() 
        tn = ((y_pred < t) & (y_val == 0)).sum() 
        scores.append((t, tp, fp, fn, tn)) 
        
        df_scores = pd.DataFrame(scores) # turn list to a pandas dataframe
    
    df_scores.columns = ['threshold', 'tp', 'fp', 'fn', 'tn'] # assign names to the columns of the dataframe

    # add tpr and fpr columns to the dataFrame, with their respective values.
    df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
    df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)

    return df_scores


scores = [] 
    
thresholds = np.linspace(0, 1, 101) 

for t in thresholds: 
    tp = ((y_pred >= t) & (y_val == 1)).sum() 
    fp = ((y_pred >= t) & (y_val == 0)).sum() 
    fn = ((y_pred < t) & (y_val == 1)).sum() 
    tn = ((y_pred < t) & (y_val == 0)).sum() 
    scores.append((t, tp, fp, fn, tn)) 
        
df_scores = pd.DataFrame(scores) # turn list to a pandas dataframe
    
df_scores.columns = ['threshold', 'tp', 'fp', 'fn', 'tn'] # assign names to the columns of the dataframe

# add tpr and fpr columns to the dataFrame, with their respective values.
df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)

# plt.plot(df_scores.threshold, df_scores.tpr, label='TPR')
# plt.plot(df_scores.threshold, df_scores.fpr, label='FBR')
# plt.legend()

#To better understand what these TPR and FPR mean, letâ€™s compare it with two baseline models: a random model and the ideal model.

#random model: outputs a random score between 0 and 1, regardless of the input.
np.random.seed(1) 
y_rand = np.random.uniform(0, 1, size=len(y_val)) 
df_rand = tpr_fpr_dataframe(y_val, y_rand)
df_rand[::10]

# plt.plot(df_rand.threshold, df_rand.tpr, label='TPR')
# plt.plot(df_rand.threshold, df_rand.fpr, label='FBR')
# plt.legend()

#ideal model: we Assume that this model always make correct decisions(churn).
num_neg = (y_val == 0).sum()
num_pos = (y_val == 1).sum()

y_ideal = np.repeat([0, 1], [num_neg, num_pos])
y_pred_ideal = np.linspace(0, 1, num_neg + num_pos)

df_ideal = tpr_fpr_dataframe(y_ideal, y_pred_ideal)
df_ideal[::10]


plt.plot(df_ideal.threshold, df_ideal.tpr, label='TPR')
plt.plot(df_ideal.threshold, df_ideal.fpr, label='FBR')
plt.legend()

# ROC Curve
plt.figure(figsize=(5, 5)) # make the plot square

plt.plot(df_scores.fpr, df_scores.tpr)
plt.plot([0, 1], [0, 1]) #random is always a straight line, so no need to generate it every time to use ROC Curve
plt.legend()

#instead of computing all the above manually (fpr, tpr, thresholds), we can use siciktLearn metrics package
fpr, tpr, thresholds = roc_curve(y_val, y_pred)

plt.figure(figsize=(5, 5))
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1])

# comparing multiple models (small, large models) to see which one will be near the ideal point(0,1) as much as we can. (TPR => y-axis, FPR => x-axis)
# Plotting multiple ROC curves on the same graph helps us visually identify which model performs better.
fpr_large, tpr_large, _ = roc_curve(y_val, y_pred)
fpr_small, tpr_small, _ = roc_curve(y_val, y_pred_small)

plt.figure(figsize=(5, 5))
plt.plot(fpr_large, tpr_large, color='black', label='Large')
plt.plot(fpr_small, tpr_small, color='green', label='Small')
plt.plot([0, 1], [0, 1])
plt.legend()

# AUC(area under the ROC curve)
# An AUC of 0.9 is indicative of a reasonably good model; 0.8 is OK, 0.7 is not very performant, and 0.6 indicates quite poor performance.
# Churn prediction is a complex problem, so an AUC of 80% is quite good.
# auc(df_scores.fpr, df_scores.tpr) #large model

#if all we need is the AUC, we can use a shortcut in siciktLearn, which will handel all of that for us, which is roc_auc_score function 
roc_auc_score(y_val, y_pred)
roc_auc_score(y_val, y_pred_small)

#AUC has a very nice probabilistic interpretation: it tells us what the probability is that a randomly selected positive example will have a score higher than a randomly selected negative example.
# we can verify that manually
neg = y_pred[y_val == 0]
pos = y_pred[y_val == 1]

np.random.seed(1)
neg_choice = np.random.randint(low=0, high=len(neg), size=10000)
pos_choice = np.random.randint(low=0, high=len(pos), size=10000)
# gives us a very close value to roc_auc_score(y_val, y_pred), due to sicikt-learn use some perscision ways that's more good, than handling it manually
(pos[pos_choice] > neg[neg_choice]).mean()

# 10- Parameter tuning

# k-fold cross validation for parameter tuning.
# simplifing the training process(for reuse) to be easy to run k-fold cross validation multiple times on it.
def train(df, y, C):
    cat = df[categorical + numerical].to_dict('records')
    dv = DictVectorizer(sparse=False) 
    dv.fit(cat) 
    X = dv.transform(cat) 
    model = LogisticRegression(solver='liblinear', C=C) #handling regularization using c parameter (to select the best parameter using regularization)
    model.fit(X, y) 
    return dv, model

def predict(df, dv, model):
    cat = df[categorical + numerical].to_dict('records') 
    X = dv.transform(cat) 
    y_pred = model.predict_proba(X)[:, 1] 
    return y_pred

nfolds = 10
kfold = KFold(n_splits=nfolds, shuffle=True, random_state=1)

# aucs = []

# for train_idx, val_idx in kfold.split(df_train_full):
#     df_train = df_train_full.iloc[train_idx].reset_index(drop=True)
#     df_val = df_train_full.iloc[val_idx].reset_index(drop=True)
    
#     y_train = df_train.churn.values
#     y_val = df_val.churn.values
    
#     dv, model = train(df_train, y_train, c=c)
#     y_pred = predict(df_val, dv, model)
    
#     auc = roc_auc_score(y_val, y_pred)
#     aucs.append(auc)

# we get 10 AUC scores evalutaed on 10 different validation folds
# Now, not only we know the average performance, but we also have an idea of how volatile
# that performance is, or how far it may deviate from the average (standard deviation).
# print('auc = %0.3f + %0.3f' % (np.mean(aucs), np.std(aucs)))

# Finding best parametes using K-fold cross-validation
# Tuning the model: selecting the best parameter C using cross-validation

for c in [0.001, 0.01, 0.1, 0.5, 1, 10]:
    aucs = []
    
    for train_idx, val_idx in kfold.split(df_train_full):
        df_train = df_train_full.iloc[train_idx]
        df_val = df_train_full.iloc[val_idx]

        y_train = df_train.churn.values
        y_val = df_val.churn.values

        dv, model = train(df_train, y_train, c)
        y_pred = predict(df_val, dv, model)

        auc = roc_auc_score(y_val, y_pred)
        aucs.append(auc)
        
    print('c=%s, auc = %0.3f Â± %0.3f' % (c, np.mean(aucs), np.std(aucs)))
    
# we will use c=0.5 as auc after c=1 doesn't grow anymore, we used 0.5 specificaly because when the c parameter is small, the model is more regularized.
# Note: The weights of this model are more restricted, so in general they are smaller. 
# Small weights in the model give us additional assurance that the model will behave well when we use it on real data. 

# 11- train the model on the entire train and validation datasets and apply it to the test dataset to verify it indeed works well.
y_train = df_train_full.churn.values
y_test = df_test.churn.values

dv, model = train(df_train_full, y_train, 0.5)
y_pred = predict(df_test, dv, model)

auc = roc_auc_score(y_test, y_pred)
print('auc = %.3f' % auc)
# we get 0.858, the important thing is that score is not significantly different from the validation score.

# probability of churning for a single customer
customer = {
    'customerid': '8879-zkjof',
    'gender': 'female',
    'seniorcitizen': 0,
    'partner': 'no',
    'dependents': 'no',
    'tenure': 41,
    'phoneservice': 'yes',
    'multiplelines': 'no',
    'internetservice': 'dsl',
    'onlinesecurity': 'yes',
    'onlinebackup': 'no',
    'deviceprotection': 'yes',
    'techsupport': 'yes',
    'streamingtv': 'yes',
    'streamingmovies': 'yes',
    'contract': 'one_year',
    'paperlessbilling': 'yes',
    'paymentmethod': 'bank_transfer_(automatic)',
    'monthlycharges': 79.85,
    'totalcharges': 3320.75,
    }

#it gives a probability of 6% that this user will churn.
# df = pd.DataFrame([customer])
# y_pred = predict(df, dv, model)
# y_pred[0]


def predict_single(customer, dv, model): 
    X = dv.transform([customer]) 
    y_pred = model.predict_proba(X)[:, 1] 
    return y_pred[0] 

predict_single(customer, dv, model)

# 11- saving the state of the model

# Pickle is a serialization/deserialization module thatâ€™s already built-in in Python: using it we can
# save an arbitrary Python object (with a few exceptions) to a file. Once we have a file, we can
# load the model from there in a different process.

with open('churn-model.bin', 'wb') as f_out:
    pickle.dump((dv, model), f_out)

# 12- loading the model

with open('churn-model.bin', 'rb') as f_in:
    dv, model = pickle.load(f_in)
    
```
