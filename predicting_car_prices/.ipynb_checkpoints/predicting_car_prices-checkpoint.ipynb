```python
import pandas as pd 
import numpy as np 

from matplotlib import pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
df = pd.read_csv('data.csv')
```

```python
#Cleaning Data:
#replacing all spaces with underscore
df.columns = df.columns.str.lower().str.replace(' ', '_') 

string_columns = list(df.dtypes[df.dtypes == 'object'].index) 

for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_') 
```

```python
sns.distplot(df.msrp[df.msrp < 100000], kde=False)
log_price = np.log1p(df.msrp) # numpy
sns.displot(log_price, kde=False)
df.isnull().sum() # pandas
```

```python

# prepering model before training.

#getting whole data in data set
n = len(df)

#make the data split into parts for training, validation, testing
n_val = int(0.2 * n)
n_test = int(0.2 * n)
n_train = n - (n_val + n_test)

#shuffled array
np.random.seed(2) #psudorandom
idx = np.arange(n)
np.random.shuffle(idx)

#creating suffled dataframe using shuffled array
df_shuffled = df.iloc[idx]
#shuffled dataframe
df_train = df_shuffled.iloc[:n_train].copy()
df_val = df_shuffled.iloc[n_train:n_train+n_val].copy()
df_test = df_shuffled.iloc[n_train+n_val:].copy()

#removing outliers using log(), we separate the actual prices for each part of our data to compare it to the predictions of that prices that the model will generate
y_train = np.log1p(df_train.msrp.values)
y_val = np.log1p(df_val.msrp.values)
y_test = np.log1p(df_test.msrp.values)

# remove prices from the dataset
del df_train['msrp']
del df_val['msrp']
del df_test['msrp']

# if 
def linear_regression(X, y):
    # adding the dummy column
    ones = np.ones(X.shape[0]) 
    X = np.column_stack([ones, X]) 
    # normal equation formula
    XTX = X.T.dot(X) 
    XTX_inv = np.linalg.inv(XTX) 
    w = XTX_inv.dot(X.T).dot(y) 
    return w[0], w[1:] 

base = ['engine_hp', 'engine_cylinders', 'highway_mpg', 'city_mpg',
'popularity']

# preparing data by filling missing value with zero and handling categorical data, while adding them to the feature set .
def prepare_x(df):
    df.copy()
    features = base.copy()
    #elementwise operation, as We calculate the difference between 2017 and each element of the year series.
    df['age'] = 2017 - df.year 
    features.append('age')
    
    # handling categorical variables for car doors
    for v in [2,3,4]:
        feature = 'num_doors_%s' % v
        value = (df['number_of_doors'] == v).astype(int)
        df[feature] = value
        features.append(feature)
        
    # handling categorical variables for car makers as bmw, here we choose the most five appeared car makers in our dataset using: df['make'].value_counts().head(5) command
    for v in ['chevrolet', 'ford', 'volkswagen', 'toyota', 'dodge']: 
        feature = 'is_make_%s' % v
        df[feature] = (df['make'] == v).astype(int)
        features.append(feature)
        
    for v in ['regular_unleaded', 'premium_unleaded_(required)',
        'premium_unleaded_(recommended)',
        'flex-fuel_(unleaded/e85)']: 
        feature = 'is_type_%s' % v
        df[feature] = (df['engine_fuel_type'] == v).astype(int)
        features.append(feature)
    
    for v in ['automatic', 'manual', 'automated_manual']: 
        feature = 'is_transmission_%s' % v
        df[feature] = (df['transmission_type'] == v).astype(int)
        features.append(feature)
    
    for v in ['front_wheel_drive', 'rear_wheel_drive',
        'all_wheel_drive', 'four_wheel_drive']: 
        feature = 'is_driven_wheels_%s' % v
        df[feature] = (df['driven_wheels'] == v).astype(int)
        features.append(feature)
    
    for v in ['crossover', 'flex_fuel', 'luxury',
        'luxury,performance', 'hatchback']: 
        feature = 'is_mc_%s' % v
        df[feature] = (df['market_category'] == v).astype(int)
        features.append(feature)
    
    for v in ['compact', 'midsize', 'large']: 
        feature = 'is_size_%s' % v
        df[feature] = (df['vehicle_size'] == v).astype(int)
        features.append(feature)
    
    for v in ['sedan', '4dr_suv', 'coupe', 'convertible',
        '4dr_hatchback']: 
        feature = 'is_style_%s' % v
        df[feature] = (df['vehicle_style'] == v).astype(int)
        features.append(feature)
    
    df_num = df[features]
    #any missing values assign zero to it. so it doesn't affect our model predictions.
    df_num = df_num.fillna(0)
    x = df_num.values
    return x

# implementing linear regression with regularization, as to make sure that the inverse exists by forcing the matrix to be invertible by adding identity matrix and multiple it's diagnol with a small number then add that to the XTX
def linear_regression_reg(X, y, r=0.0):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    
    XTX = X.T.dot(X)
    reg = r * np.eye(XTX.shape[0]) 
    XTX = XTX + reg 
    
    XTX_inv = np.linalg.inv(XTX)
    
    w = XTX_inv.dot(X.T).dot(y)
    
    return w[0], w[1:]
    

x_train = prepare_x(df_train)

w_0, w = linear_regression_reg(x_train, y_train, r=0.001) #we use linear regression to extract weight, by giving it the training data and their prices to train on.

y_pred = w_0 + x_train.dot(w) #then multiply it with set of features to predict our prices

#plot the predicted values and compare them with the actual prices for training set
# sns.distplot(y_pred, label='prediction')
# sns.distplot(y_train, label='target')
# plt.legend()

#implementaton of root mean squared error (to quantify error of our model)
def rmse(y, y__pred):
    error = y_pred - y # array error contains differences
    mse = (error ** 2).mean()
    return np.sqrt(mse)

print('Training', rmse(y_train, y_pred))

# by using validation data, we check whether any model adjustments lead to improvements in the predictive quality of the model
x_val = prepare_x(df_val)
y_pred = w_0 + x_val.dot(w) #contains prediction for the validation dataset
print('Validation:', rmse(y_val, y_pred))


x_test = prepare_x(df_test)
y_pred = w_0 + x_test.dot(w) #contains prediction for the validation dataset
print('Test:', rmse(y_test, y_pred))


# df_train.year.max() to see the maximum year to subtract from it, to get the age of the car

#plot the predicted values and compare them with the actual prices of validation set
sns.histplot(y_pred, label='prediction')
sns.histplot(y_val, label='target')
plt.legend()




# predicting a price for an actual car

ad = {
    'city_mpg': 18,
    'driven_wheels': 'all_wheel_drive',
    'engine_cylinders': 6.0,
    'engine_fuel_type': 'regular_unleaded',
    'engine_hp': 268.0,
    'highway_mpg': 25,
    'make': 'toyota',
    'market_category': 'crossover,performance',
    'model': 'venza',
    'number_of_doors': 4.0,
    'popularity': 2031,
    'transmission_type': 'automatic',
    'vehicle_size': 'midsize',
    'vehicle_style': 'wagon',
    'year': 2013
}

df_test = pd.DataFrame([ad])
x_test = prepare_x(df_test)
y_pred = w_0 + x_test.dot(w)

suggestion = np.expm1(y_pred)
suggestion # output is equal 28294.64008002 dollars, while actuall car price is 31,120 dollars, which is not far
```
